```python
# 룩업 테이블의 생성
min_table  =  40
max_table  =  220
diff_table  =  max_table  -  min_table

LUT_HC  =  np . arange ( 256 ,  dtype  =  'uint8'  ) 
LUT_LC  =  np . arange ( 256 ,  dtype  =  'uint8'  )

# 경조 LUT 생성
for  i  in  range ( 0 ,  min_table ) : 
    LUT_HC [ i ]  =  0 
for  i  in  range ( min_table ,  max_table ) : 
    LUT_HC [ i ]  =  255  *  ( i  -  min_table )  /  diff_table 
for  i  in  range ( max_table ,  255 ) : 
    LUT_HC [ i ]  = 255

# 로우 콘트라스트 LUT 생성
for  i  in  range ( 256 ) : 
    LUT_LC [ i ]  =  min_table  +  i  *  ( diff_table )  /  255
   
LUTs = []
LUTs . append ( LUT_HC )
LUTs . append ( LUT_LC )    
    
    

for a in range (100):
    img_src  =  cv2 . imread ("onion/onion_" + str(a + 101) + ".jpg", 1 ) 

    

    
    for  i , LUT  in  enumerate ( LUTs ) :
        img_src = ( cv2 . LUT ( img_src , LUT ))
    
    

    # 저장
    if  not  os . path . exists ( "onion_images" ) :  # 폴더 생성
        os . mkdir ( "onion_images" )
        
    pil_img  =  Image. fromarray ( img_src ) 
    pil_img . save ( "onion_images/onion_" + str(a + 1) + ".jpg" ) 


  
```

```python
gamma1  =  0.75 
gamma2  =  1.5 
gamma3  =  0.5
gamma4  =  1.25
for  i  in  range ( 256 ) : 
    LUT_G1 [ i ]  =  255  *  pow ( float ( i )  /  255 ,  1.0  /  gamma1 ) 
    LUT_G2 [ i ]  =  255  *  pow ( float ( i )  /  255 ,  1.0  /  gamma2 )   
    LUT_G3 [ i ]  =  255  *  pow ( float ( i )  /  255 ,  1.0  /  1 )
    LUT_G4 [ i ]  =  255  *  pow ( float ( i )  /  255 ,  1.0  /  gamma3 ) 
    LUT_G5 [ i ]  =  255  *  pow ( float ( i )  /  255 ,  1.0  /  gamma4 ) 
```
